```python
import spacy
from spacy.lang.pt.stop_words import STOP_WORDS
import string

# Ativar a GPU
spacy.require_gpu()

# Carregar o modelo de português
nlp = spacy.load("pt_core_news_lg")
```

```python
if spacy.prefer_gpu():
    print("Usando GPU!")
else:
    print("Usando CPU.")
```

```python
# Criar lista de stopwords e pontuação
stopwords = list(STOP_WORDS)
pontuacao = string.punctuation

# Processar o texto com o spaCy
doc = nlp(texto)

# Remover stopwords e pontuação antes de lematizar
tokens_sem_stopwords = [token for token in doc if token.text.lower() not in stopwords and token.text not in pontuacao]

# Lematizar os tokens restantes
texto_lemmatizado = " ".join([token.lemma_ for token in tokens_sem_stopwords])

# Exibir o texto original e o texto lematizado
print(f"Texto original: {texto}")
print(f"Texto lematizado: {texto_lemmatizado}")
```

```python
from enelvo.normaliser import Normaliser
norm = Normaliser(tokenizer='readable')
```

```python
# Aplicar o stemmer para cada palavra do texto e reconstruir o texto
texto_normalizado = " ".join([norm.normalise(palavra) for palavra in texto.split()])

# Exibir o texto original e o texto stemmizado
print(f"Texto original: {texto}")
print(f"Texto normalizado: {texto_normalizado}")
```

```python
import stanza

stanza.download('pt')
nlp = stanza.Pipeline('pt')
```

```python
# Inicializar o pipeline para português
nlp_stanza = stanza.Pipeline('pt', processors='tokenize,mwt,pos,lemma')

# Processar o texto com o Stanza
doc_stanza = nlp_stanza(texto)

# Extrair os lemas e reconstruir o texto lematizado
texto_lemmatizado_stanza = " ".join([word.lemma for sent in doc_stanza.sentences for word in sent.words])

# Exibir o texto original e o texto lematizado
print(f"Texto original: {texto}")
print(f"Texto lematizado: {texto_lemmatizado_stanza}")
```
